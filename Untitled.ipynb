# Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
display(df.describe())
df.info()
```

### transform from square feet to meters by: sqft/10.764

```python
df['meters_living'] = df['sqft_living']/10.764
df['meters_lot'] = df['sqft_lot']/10.764
df['meters_above'] = df['sqft_above']/10.764
df['meters_basement'] = df['sqft_basement']/10.764
```

```python
df = df.drop(['sqft_lot15', 'sqft_living15', 'zipcode', 'id', 'date', 'yr_renovated', 'sqft_living', 'sqft_lot', 'sqft_above', 'sqft_basement'], axis=1)
df.head()
```

```python
df.to_csv('cleaner_data.csv')
```

# Machine Learning

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
dtr = DecisionTreeRegressor()
lr = LinearRegression()
```

```python
x = df.drop('price', axis=1)
y = df['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
y_test
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
r2_score(y_test, y_pred)
```
