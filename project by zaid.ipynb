### import essential libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### the dataset

https://www.kaggle.com/datasets/arathipraj/house-data

```python
df = pd.read_csv('house_data.csv')
df.head()
```

```python
print(df.info())
display(df.describe())
print(df.isnull().sum())
```

```python
df = df.drop(['id', 'date'], axis=1)
```

## machine learning stuff

### import machine learning libraries

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

from sklearn.metrics import r2_score
```

### split data into x and y

```python
x = df.drop(['price'], axis=1)
y = df['price']
```

### build models

```python
lr = LinearRegression()
dtr = DecisionTreeRegressor()
rfr = RandomForestRegressor()
svr = SVR()
```

### compare the models and select the best one

```python
list_of_metrics = []
iterations = 10
for i in range(1, iterations+1):
    x_train, x_test, y_train, y_test = train_test_split(x,y)
    
    lr.fit(x_train, y_train)
    dtr.fit(x_train, y_train)
    rfr.fit(x_train, y_train)
    svr.fit(x_train, y_train)

    lr_predictions = lr.predict(x_test)
    dtr_predictions = dtr.predict(x_test)
    rfr_predictions = rfr.predict(x_test)
    svr_predictions = svr.predict(x_test)
    
    list_of_metrics.append([
        r2_score(y_test, lr_predictions),
        r2_score(y_test, dtr_predictions),
        r2_score(y_test, rfr_predictions),
        r2_score(y_test, svr_predictions)]
    )
    print(f'done with {i} iterations')
```

```python
r2 = pd.DataFrame(list_of_metrics, columns=['Linear Regression', 'Decision Tree', 'Random Forest', 'SVR'])
```
